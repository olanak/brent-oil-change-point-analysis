```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Load processed data
df = pd.read_csv('../data/Processed/processed_data.csv', parse_dates=['Date'], index_col='Date')
```

```python
# Plot raw prices
plt.figure(figsize=(12, 6))
plt.plot(df['Price'], label='Brent Oil Price', color='blue')
plt.title('Historical Brent Oil Prices (19872022)')
plt.xlabel('Date')
plt.ylabel('Price (USD/barrel)')
plt.legend()
plt.grid(True)
plt.savefig('../docs/price_trend.png')  # Save for reporting
plt.show()
```

```python
# Plot log returns for stationarity check
plt.figure(figsize=(12, 4))
plt.plot(df['Log_Return'], label='Daily Log Returns', color='green')
plt.title('Log Returns (Stationarity Check)')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.legend()
plt.grid(True)
plt.savefig('../docs/log_returns.png')
plt.show()
```

```python
# ACF/PACF plots for ARIMA parameter selection
plt.figure(figsize=(12, 6))
plot_acf(df['Log_Return'].dropna(), lags=30, alpha=0.05, title='ACF of Log Returns')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plot_pacf(df['Log_Return'].dropna(), lags=30, alpha=0.05, title='PACF of Log Returns')
plt.show()
```
